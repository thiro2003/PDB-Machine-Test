# • Handle missing values using two different techniques

```python
# • Handle missing values using two different techniques
# ANS:- 

# let's find the missing values 
data.isnull().sum()
```

```python
# We have the following missing value
# Age            2 missing
# Gender         2 missing
# Salary         2 missing 
```

```python
# Let's handle each individualy
# 1.Missing values in Age 
# :-
# let check the missing values 
data[data.isnull().any(axis=1)]
```

```python
# As we can see above there are 2 missing values which shows NaN 
# So we can either remove or replace the values but i would choose to replace it because we have small dataset so removing entire column
# can effect the model performace so i gonna replace it with mean of the the all ages this will disolve the in the dataset and it will not effect my ML model
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
```

```python
# let see missing age replaced or not 
data.isnull().sum()
```

```python
# As we can see above values has been replaced with mean 
# let's take gender
data[data.isnull().any(axis=1)]
```

```python
# since gender is in binary value and in text it is not possible to calculate mean so can use Mode(most freq occuring value) but it is random and it will change the shape of data distribution so  i'll apply feature enginnering 
#  feature engineering the process of the applying bussiness knowledge and getting the most aproximate data to train the model.
# We can see that in Gender's missing value  we have name so we can deteremine the gender using name

# I below row i can clearly see that name of employee is sneha so it is likely that she is Female 
# 3	Sneha Kulkarni	30.0	NaN	48000.0	Finance	22-07-2020

# I below row i can clearly see that name of employee is sneha so it is likely that she is male 
# 10	Karan Shah	36.0	NaN	70000.0	Operations	30-05-2016

# So at the end i will repplace the Sneha's gender with female and Karan's gender with the male
```

```python
data.loc[data['Name'] == 'Sneha Kulkarni', 'Gender'] = 'Female'
data.loc[data['Name'] == 'Karan Shah', 'Gender'] = 'male'
```

```python
data.isnull().sum()
```

```python
# As we can see that above we have only salary column with missing value
data[data.isnull().any(axis=1)]
```

```python
# I'll use the same method i've used on the age column our main goal is to keep data in the dataset to make data consistency but at the same time model accuracy . 
# let's replace missing salary with the mean of all saries 
data['Salary']=data['Salary'].fillna(data['Salary'].mean())
```

# • Detect outliers using IQR method

```python
data.head()
```

```python
# To check outlier so we can only check quatitative values such as (Age, Salary) 
# Outliers are extreme most or least value it depend on the dataset 
# :-
# formula to calculate outlier using IQR is IQR(Inter Quartile Range) =Q3-Q1 so we have calculate quantile 

# Step-1:- Calculate Quantiles 
# Step-2:- Putting them in the formula 
# Step-3:- calculating lower and upper bound 
# Step-4:- declaring value below lower bound is outlier and value high than upper bound 
```

```python
# checking outlier for age 
Q1=data['Age'].quantile(0.25)
Q3=data['Age'].quantile(0.75)

IQR=Q3-Q1 
lower_bound=Q1-1.5*IQR 
upper_bound=Q3+1.5*IQR

outliers = data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)]
outliers
```
